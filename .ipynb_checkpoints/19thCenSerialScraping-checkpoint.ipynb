```python
def getsourcekey(key, item):
    tempdict=item['docs'][0]['source_record']
    if tempdict.has_key(key):
        temp=tempdict[key]
    else:
        temp=""
    if temp=="":
        return "NA"
    else:
        return temp
```

```python
df=pd.DataFrame()
```

```python
[getsourcekey("245a",dummy) for dummy in listofdicts][:3]
```

```python
for key in ['245a', '362a', '525a']:
    df[key] = [getsourcekey(key, item) for item in listofdicts]
```
